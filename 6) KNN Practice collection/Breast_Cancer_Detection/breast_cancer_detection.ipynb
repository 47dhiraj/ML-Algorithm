**Splitting the dataset into 75% training & 25% testing**

```python
from sklearn.model_selection import train_test_split            # importing train_test_split from sklearn
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, test_size=0.2)
```

**Scaling the Feature (Feature Scaling)**

**The main reason to scaling the feature is that after scaling the model gives the best output**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# print(X_train)
```

```python
# print(X_test)
```

**Create a function for the Models**

```python
def models(X_train, Y_train):
    
    # Creating Logistic Regression Model
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(X_train, Y_train)
    
    # Creating Decision Tree Model
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
    tree.fit(X_train, Y_train)
    
    # Creating Random Forest Classifier Model
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10, criterion='entropy', random_state=0)
    forest.fit(X_train, Y_train)
    
    # Creating K-Nearest Neighbours Classifier Model
    from sklearn.neighbors import KNeighborsClassifier
    knnclassifier = KNeighborsClassifier(n_neighbors=11, p=2,metric='euclidean')
    knnclassifier.fit(X_train, Y_train)
    
    
    #Print the model accuracy on the training data
    print('Logistic Regression Training Accuracy:', log.score(X_train, Y_train))
    print('Decision Tree Training Accuracy:', tree.score(X_train, Y_train))
    print('Random Forest Classifier Training Accuracy:', forest.score(X_train, Y_train))
    print('KNN Classifier Training Accuracy:', knnclassifier.score(X_train, Y_train))
    
    
    return log, tree, forest, knnclassifier
```

**Getting all of this models**

```python
model = models(X_train, Y_train)\
```

**Evaluating model accuracy on the testing data**

```python
from sklearn.metrics import confusion_matrix                      # importing confusion_matrix
```

```python
for i in range(len(model)):
    cm = confusion_matrix(Y_test, model[i].predict(X_test)) 
    print('For Model', i)
    print(cm)

    TruePositive = cm[0][0]
    TrueNegative = cm[1][1]
    FalseNegative = cm[1][0]
    FalsePositive = cm[0][1]

    print('TruePositive :',TruePositive)
    print('TrueNegative :',TrueNegative)
    print('FalseNegative :',FalseNegative)
    print('FalsePositive :',FalsePositive)
    
    print('Testing Accuracy =', (TruePositive + TrueNegative)/ (TruePositive + TrueNegative + FalseNegative + FalsePositive), '\n')
```

**Using sklearn inbuilt method to get the metrices of the models**

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
for i in range(len(model)):
    print('For Model', i)
    print(classification_report(Y_test, model[i].predict(X_test)))
    print('Testing Accuracy Score of the Model',[i],' :',accuracy_score(Y_test, model[i].predict(X_test)), '\n\n')
```
